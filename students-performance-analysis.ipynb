## About Dataset
### Context
Marks secured by the students

### Content
This data set consists of the marks secured by the students in various subjects.

### Inspiration
To understand the influence of the parents background, test preparation etc on students performance

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
warnings.warn("this will not show")
```

```python
df = pd.read_csv("/kaggle/input/students-performance-in-exams/StudentsPerformance.csv") 
df
```

```python
df.info() 
```

```python
df.describe().T 
```

```python
plt.figure(figsize=(15, 5))

# Math
plt.subplot(1, 3, 1)
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Mathematics Achievement by Gender')

# Reading
plt.subplot(1, 3, 2)
sns.boxplot(x='gender', y='reading score', data=df)
plt.title('Reading Achievement by Gender')

# Writing
plt.subplot(1, 3, 3)
sns.boxplot(x='gender', y='writing score', data=df)
plt.title('Writing Success by Gender')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.boxplot(x='parental level of education', y='math score', data=df)
plt.title('Mathematics Achievement According to Parental Education Level')
plt.xticks(rotation=45)

plt.subplot(1, 3, 2)
sns.boxplot(x='parental level of education', y='reading score', data=df)
plt.title('Reading Achievement According to Parental Education Level')
plt.xticks(rotation=45)

plt.subplot(1, 3, 3)
sns.boxplot(x='parental level of education', y='writing score', data=df)
plt.title('Writing Achievement According to Parent Education Level')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 5))

# Math Achievement
plt.subplot(1, 3, 1)
sns.boxplot(x='lunch', y='math score', data=df)
plt.title('Mathematics Achievement by Food Status')
plt.xticks(rotation=45)

# Reading Achievement
plt.subplot(1, 3, 2)
sns.boxplot(x='lunch', y='reading score', data=df)
plt.title('Reading Achievement According to Food Status')
plt.xticks(rotation=45)

# Writing Success
plt.subplot(1, 3, 3)
sns.boxplot(x='lunch', y='writing score', data=df)
plt.title('Writing Success by Food Situation')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 5))


plt.subplot(1, 3, 1)
sns.barplot(x='test preparation course', y='math score', data=df)
plt.title('Mathematics Achievement by Preparatory Course')


plt.subplot(1, 3, 2)
sns.barplot(x='test preparation course', y='reading score', data=df)
plt.title('Reading Achievement by Preparatory Course')


plt.subplot(1, 3, 3)
sns.barplot(x='test preparation course', y='writing score', data=df)
plt.title('Writing Success According to Preparatory Course')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.barplot(x='race/ethnicity', y='math score', data=df)
plt.title('Mathematics Achievement by Ethnicity')

plt.subplot(1, 3, 2)
sns.barplot(x='race/ethnicity', y='reading score', data=df)
plt.title('Reading Achievement by Ethnicity')

plt.subplot(1, 3, 3)
sns.barplot(x='race/ethnicity', y='writing score', data=df)
plt.title('Writing Success by Ethnicity')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.pointplot(x='parental level of education', y='math score', hue='test preparation course', data=df, 
              dodge=True, markers=["o", "s"], linestyles=["-", "--"])
plt.title('Interaction Between Parent Education and Test Prep Course')
plt.xlabel('Parent Education')
plt.ylabel('Math Score')
plt.show()
```

### Shaping Equitable Education Policies: Next Steps

It is critical to acknowledge that a student's success is shaped by both academic skills and external factors, like parental education and access to test prep. To level the playing field, we must implement education policies that:

Promote Tailored Support: Develop inclusive programs that offer customized support based on individual student needs and backgrounds, considering factors such as parental education.
Bridge Resource Gaps: Expand access to vital resources like test preparation courses, especially for students from disadvantaged backgrounds, to reduce disparities in opportunity and achievement.
Ensure Equal Access: Strive to make equal educational opportunities a reality for all students, regardless of their socioeconomic background or parental education level.
